```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.loc[    df["koi_disposition"]=="FALSE POSITIVE", "koi_disposition"] ="FALSEPOSITIVE"
df.head()
```

# Select your features (columns)

```python
print(df)
```

```python
# Set features. This will also be used as your x values.
X = df.drop(columns=["koi_disposition"])
print(X)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df["koi_disposition"]
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

label_encoder.fit(y)
encodeYTrain = label_encoder.transform(y_train)
encodeytest = label_encoder.transform(y_test)
```

```python
from keras.utils import to_categorical
final_y_train = to_categorical(encodeYTrain)
final_y_train
```

```python
from keras.utils import to_categorical
final_y_test = to_categorical(encodeytest)
final_y_test
```

# Train the Model



```python
from tensorflow.keras.models import Sequential

model = Sequential()
```

```python
from tensorflow.keras.layers import Dense
model.add(Dense(units=100,activation='relu', input_dim=40))
model.add(Dense(units=100,activation='relu'))
model.add(Dense(units=3,activation='softmax'))
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Fit (train) the model
model.fit(
    X_train_scaled,
    final_y_train,
    epochs=50,
    shuffle=True,
    
)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

```python
loss, acc = model.evaluate(X_train_scaled, final_y_train)
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model

from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, final_y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
