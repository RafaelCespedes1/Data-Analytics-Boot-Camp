```python
census_combined_pd["central_americans"] = census_combined_pd["nicaraguans"]+census_combined_pd["costa_ricans"]+census_combined_pd["guatemalans"]+census_combined_pd["hondurans"]+census_combined_pd["panamanians"]+census_combined_pd["salvadorans"]
```

```python
census_combined_pd["total"] = census_combined_pd["central_americans"]+census_combined_pd["south_americans"]
```

```python
census_combined_pd["percent_central_american"] = 100 * census_combined_pd["central_americans"] / census_combined_pd["population"]
census_combined_pd["percent_south_american"] = 100 * census_combined_pd["south_americans"] / census_combined_pd["population"]
census_combined_pd["percent_total"] = 100 * census_combined_pd["total"] / census_combined_pd["population"]
```
