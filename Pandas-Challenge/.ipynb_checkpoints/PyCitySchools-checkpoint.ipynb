```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
a=len(school_data_complete["school_name"].unique())
print("The total number of schools is " +str(a))
# Calculate the total number of students
b=len(school_data_complete["Student ID"].unique())
print("The total number of students is " +str(b))
# Calculate the total budget
c = school_data_complete["budget"].unique().sum()
print("The total budget is " +str(c))
# Calculate the average math score
d = school_data_complete["math_score"].mean()
print("The average math score is " +str(d))
# Calculate the average reading score
e = school_data_complete["reading_score"].mean()
print("The average reading score is " +str(e))
# Calculate the percentage of students with a passing math score (70 or greater)
f = len(school_data_complete[school_data_complete["math_score"] >= 70])
print("The total number of students with a passing math score (70 or greater) is " +str(f))
print("The percentage of students with a passing math score (70 or greater) is " +str(100*f/b))
# Calculate the percentage of students with a passing reading score (70 or greater)
g = len(school_data_complete[school_data_complete["reading_score"] >= 70])
print("The total number of students with a passing reading score (70 or greater) is " +str(g))
print("The percentage of students with a passing reading score (70 or greater) is " +str(100*g/b))
# Calculate the percentage of students who passed math and reading (% Overall Passing)
h = len(school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)])
print("The total number of students with an overall passing score (70 or greater) is " +str(h))
print("The percentage of students with an overall passing score (70 or greater) is " +str(100*h/b))
school_data_District_Summary = {"Total Schools" : [a],
                                "Total Students" : [b], 
                                "Total Budget" : [c], 
                                "Average Math Score" : [d], 
                                "Average Reading Score" : [e], 
                                "% Passing Math" : [100*f/b], 
                                "% Passing Reading" : [100*g/b],
                                "% Overall Passing" : [100*h/b] }
pd.DataFrame(school_data_District_Summary)
#school_data_District_Summary().style.format({"Total Students": "{:,}", "Total Budget": "${:,.2f}"})
```

```python
school_data_complete["passed_reading"] = np.where(school_data_complete["reading_score"]>=70,1,0)
school_data_complete["passed_math"] = np.where(school_data_complete["math_score"]>=70,1,0)  
school_data_complete["passed_both"] = np.where((school_data_complete["math_score"]>=70)&(school_data_complete["reading_score"]>=70),1,0)
```
